# 1.Basic Plot

```python
#Basic Plotting: plot
#This functionality on Series and DataFrame is just a simple wrapper around the matplotlib libraries plot() method.
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/1/2000',periods=10), columns=list('ABCD'))
print(df)
df.plot()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,2),index=pd.date_range('1/1/2000',periods=10), columns=list(['A','D']))
print(df)
df.plot()
```

# 2.Bar Plot

```python
#Bar Plot with negative values

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar()
```

```python
#Bar Plot with positive values

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar()
```

```python
#To produce a stacked bar plot, pass stacked=True −

import pandas as pd
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar(stacked=True)
```

```python
#To get horizontal bar plots, use the barh method −
#stacked=true
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.barh(stacked=True)      
```

```python
#stacked=false

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.barh(stacked=False)    
```

# 3.HISTOGRAMS

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
print(df)
df.plot.hist(bins=20)
```

# 4.BAR PLOTS

```python
# boxplot representing five trials of 10 observations of a uniform random variable on [0,1)

import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
print(df)
df.plot.box()
```

# 5.AREA PLOT

```python
#Area plot can be created using the Series.plot.area() or the DataFrame.plot.area() methods

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
print(df)
df.plot.area()
```

# 6.SCATTER PLOT

```python
#Scatter plot can be created using the DataFrame.plot.scatter() methods

import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
print(df)
df.plot.scatter(x='a', y='b')
```

# 7.PIE PLOT

```python
#Pie chart can be created using the DataFrame.plot.pie() method

import pandas as pd
import numpy as np

df = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])

print(df)
df.plot.pie(subplots=True)
```
